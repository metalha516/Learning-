# Kadan's Algorithm
Use this Algorhtm Whenever you see:
1. Maximum subarray
2. Largest contiguous sum
3. Continuous segment
4. Best time window
5. Maximum profit in a continuous period
6. Find maximum sum of consecutive elements

Don’t use Kadane if:

1. Subsequence (not contiguous)
2. Need exact sum = K
3. Need number of subarrays
4. Need length constraint (ulke size exactly K)
5. It invulves sorting
6. It invulves combinations


```python
def kahans(arr):
    currentSum = 0
    MaxSum = 0
    for i in arr:
        currentSum += i
        MaxSum = max(currentSum, MaxSum)
        if currentSum < 0:
            currentSum = 0
    return MaxSum

print(kahans([2, 6, 7, 8, 3]))
```

# Moore's Voting Algorithm 

When Sould We use it : 
1. Majority Element.
2. More than half of the array.
3. O(1) Space restriction

When We Should not use it:
1. When no majority Guranted.
2. When Freequency is something else.
3. When we need's to count freequency


```python
def MooresVoting(arr):
    freq = 0
    ans = 0
    for i in arr:
        if freq == 0:
            ans = i
        if ans == i:
            freq+=1
        else:
            freq-=1
    freq = 0
    for i in arr:
        if i == ans:
            freq+=1
            
    return ans if freq > len(arr)/2 else -1
arr = [1, 3, 3, 3, 3, 5, 3, 8, 4, 3]
print(f"More than n/2 element is : {MooresVoting(arr)}")
```
